```python
import pandas as pd
import numpy as np
```

## 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
df = pd.read_csv('Data/OnlineRetail.csv', encoding='iso-8859-2')
df.head()
```

```python
print(f'Dữ liệu có {df.shape[0]} dòng')
print(f'Dữ liệu có {df.shape[1]} cột')
```

```python
df.dtypes
```

## 2. Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 

```python
pivot_table1 = df.pivot_table(values = 'Quantity', index = 'InvoiceNo', columns = 'Country', aggfunc = 'mean')
pivot_table1
```

## 3. Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

```python
pivot_table2 = df.pivot_table(values = 'Quantity', index = 'CustomerID', aggfunc = ['min', 'max'])
pivot_table2
```

## 4. Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.

```python
pivot_table3 = df.pivot_table(values = ['Quantity','UnitPrice'],
                              index = 'StockCode', aggfunc = {'Quantity':np.sum,'UnitPrice':np.mean})
pivot_table3
```

## 5. Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.

```python
df['InvoiceDate'] = df['InvoiceDate'].str.extract(r'(\d+\/\d+\/\d+)')
```

```python
pivot_table4 = df.pivot_table(values = 'Quantity',index = 'InvoiceDate', aggfunc = 'sum')
pivot_table4
```
