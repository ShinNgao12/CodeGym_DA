```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme(style='darkgrid')
```

```python
df = pd.read_csv('Data/FoodPrice_in_Turkey.csv')
df.head()
```

```python
# Lọc dữ liệu sản phẩm gạo, vẽ biểu đồ xu hướng giá qua các năm
rice_df = df[df['ProductName'] == 'Rice - Retail']
sns.lmplot(x = 'Year', y= 'Price' ,data = rice_df )
plt.title('Xu hướng giá gạo qua các năm', fontsize = 16)
plt.show()
```

 ### Giá gạo theo lmplot có xu hương tăng

```python
# Biểu đồ phân bố cho giá sản phẩm qua từng năm
sns.violinplot(x = 'Year',y = 'Price', data = rice_df)
plt.title('Phân bố  giá gạo qua các năm', fontsize = 16)
plt.show()
```

### Giá gạo theo từng năm có xu hướng tăng nhưng phân bố  càng ngày càng xa phân phối chuẩn (Violin ngày càng hẹp)

```python
# Thống kê sản phẩm theo địa điểm
sns.countplot(x = "Place", data = df)
plt.title('Thống kê số lượng sản phầm theo khu vực', fontsize = 16) # vì SL sản phẩm bán ra trừng SL khu vực xuất hiện
plt.show()
```

```python
sns.boxplot(x = "Year", y = "Price", data=df)
plt.title('Phân bố giá sản phầm theo năm', fontsize = 16)
plt.show()
```

## 2. Online Retail

```python
df = pd.read_csv('Data/OnlineRetail.csv', encoding= 'ISO-8859-1')
df.head()
```

```python
df = df.dropna()
df["Price"] = df["Quantity"] * df["UnitPrice"]
```

```python
# Vẽ biểu đồ phân bố cho giá sản phẩm
sns.violinplot(y = 'UnitPrice', data = df )
plt.title('Phân bố cho đơn giá sản phẩm', fontsize = 16)
plt.show()
```

- Dữ liệu phân bố không đều

```python
# Biểu đồ phân bố cho tổng giá mỗi sản phẩm ở các đơn
sns.violinplot(y = 'Price', data = df )
plt.title('Phân bố cho tổng giá sản phẩm', fontsize = 16)
plt.show()
```

- Dữ liệu phân bố không đều
- Có giá trị âm @@

```python
sns.boxplot(x=df["UnitPrice"])
plt.title('Phân phối đơn giá', fontsize = 16)
plt.show()
```

## 3. GDPList

```python
df = pd.read_csv('Data/GDPlist.csv')
df.head()
```

```python
# Biểu đồ phân bố giá trị GDP toàn cầu
sns.violinplot(y = 'GDP (millions of US$)', data = df)
plt.title('Phân bố giá trị GDP toàn cầu', fontsize = 16)
plt.show()
```

```python
# Biểu đồ phân bố giá trị GDP theo châu lục Á
sns.violinplot(y=df[df["Continent"]=="Asia"]["GDP (millions of US$)"])
plt.title('Phân bố giá trị GDP châu Á')
plt.show()
```

```python
sns.boxplot(x="Continent", y = "GDP (millions of US$)", data=df)
plt.title('Box plot theo châu lục')
plt.show()
```
