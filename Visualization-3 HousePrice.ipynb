```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme(style='darkgrid')
```

```python
df = pd.read_excel('Data/house_price_dống-da.xlsx')
```

```python
df.columns
```

## 1. Vẽ biểu đồ xu hướng phân tích mối quan hệ giữa giá nhà với diện tích, giá nhà với số lượng phòng ngủ, nhận xét.

```python
fig, axs = plt.subplots(1,2, figsize = (10,5))
# dùng reglot thay thế cho  lmplot khi chỉ định axis cụ thể 
sns.regplot(x = 'area', y = 'price', data = df, ax = axs[0])
sns.regplot(x = 'bedroom', y = 'price',  data = df, ax = axs[1])
plt.suptitle('Xu hướng giá nhà với diện tích và số  lượng phòng ngủ', fontsize = 16)
plt.show()
```

### Nhận xét :
- Nhìn chung, giá nhà có xu hướng tăng khi tăng diện tích hoặc số lượng phòng ngủ
- Dự vào hình vẽ ta nhận thấy một số điểm dị biệt, nên xóa bỏ
- Xu hướng của giá nhà theo số phòng hội tụ tốt hơn so với diện tích

## 2. Vẽ biểu đồ phân bố thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
plt.rcParams['figure.figsize'] = (12,5)
sns.violinplot(x = 'house_direction', y = 'price', data = df)
plt.title('Phân bố giá nhà theo hướng nhà', fontsize = 16)
plt.show()
```

### Nhận xét:
- Nhìn chung, giá nhà giữa các hướng có giá trị trung tâm gần nhau, ngoại trừ hướng `Đông - Bắc` cao hơn
- Có xuất hiện giá nhà âm ở 1 vài khu vực
- Hướng `Nam , Đông, Tây-Nam` dữ liệu có tính tập trung và khoảng tin cậy nhỏ ==> tốt
- Hướng `Băc` có gì đó không ổn ở đây

## 3. Vẽ biểu đồ tần số để đếm số nhà ở mỗi hướng nhà, nhận xét.

```python
sns.countplot(x = 'house_direction', data = df)
plt.title('Đếm số nhà theo hướng nhà', fontsize = 16)
plt.show()
```

### Nhận xét
- Hướng Đông - Nam có số lượng nhà bán cao, hướng Bắc thì thấp, cao hướng còn lại chênh lệch nhau không đáng kể

## 4. Vẽ biểu đồ boxplot thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
sns.boxplot(x = 'house_direction', y = 'price', data = df)
plt.title('Phân bố giá nhà theo hướng nhà', fontsize = 16)
plt.show()
```

- Nhìn chung, giá nhà giữa các hướng có giá trị trung tâm gần nhau, ngoại trừ hướng `Đông - Bắc` cao hơn
- Không thể phát hiện giá nhà âm như Violin plot
- Hướng `Nam , Đông, Tây-Nam` dữ liệu có tính tập trung
- Hướng `Băc` có gì đó không ổn ở đây
